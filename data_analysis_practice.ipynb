```python
def replace_empty_with_nan(data):
    if isinstance(data, dict):
        return {key: replace_empty_with_nan(value) if value != "" else np.nan for key, value in data.items()}
    elif isinstance(data, list):
        return [replace_empty_with_nan(item) for item in data]
    else:
        return data
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame
# Plot the non-null counts for each column
non_null_counts = df.notnull().sum()

plt.figure(figsize=(10, 6))
sns.barplot(x=non_null_counts.index, y=non_null_counts.values)
plt.xticks(rotation=90)
plt.xlabel('Columns')
plt.ylabel('Non-Null Count')
plt.title('Non-Null Counts of Columns')
plt.show()
```

```python
import folium
import geopandas as gpd
import pandas as pd

# Find the country with the largest impact
max_impact_country = df.loc[df['impact'].idxmax()]
print(f"Largest Impact Country: {max_impact_country['country']} with Impact Value: {max_impact_country['impact']}")

# Load world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Merge data with world map
merged = world.set_index('name').join(df.set_index('country'))

# Create a folium map
m = folium.Map(location=[20, 0], zoom_start=2)

# Function to get color based on impact
def get_color(impact):
    if impact < 0.2:
        return 'green'
    elif 0.2 <= impact < 0.4:
        return 'yellow'
    elif 0.4 <= impact < 0.6:
        return 'orange'
    else:
        return 'red'

# Add countries to the map
for _, row in merged.iterrows():
    if pd.notna(row['impact']):
        geo_json = folium.GeoJson(
            row['geometry'],
            style_function=lambda feature, impact=row['impact']: {
                'fillColor': get_color(impact),
                'color': 'black',
                'weight': 1,
                'dashArray': '5, 5',
                'fillOpacity': 0.7,
            }
        )
        geo_json.add_to(m)
        # Highlight the country with the largest impact
        if row.name == max_impact_country['country']:
            geo_json.add_child(folium.Popup(f"{row.name}: {row['impact']}"))

# If running in a Jupyter notebook, display the map
m
```

```python

# Calculate average intensity for each country
average_intensity = df.groupby('country')['intensity'].mean().reset_index()
average_intensity.columns = ['country', 'average_intensity']

# Find the country with the highest average intensity
max_intensity_country = average_intensity.loc[average_intensity['average_intensity'].idxmax()]
print(f"Highest Average Intensity Country: {max_intensity_country['country']} with Average Intensity: {max_intensity_country['average_intensity']}")

# Load world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Merge data with world map
merged = world.set_index('name').join(average_intensity.set_index('country'))

# Create a folium map
m = folium.Map(location=[20, 0], zoom_start=2)

# Function to get color based on average intensity
def get_color(intensity):
    if intensity < 0.2:
        return 'green'
    elif 0.2 <= intensity < 0.4:
        return 'yellow'
    elif 0.4 <= intensity < 0.6:
        return 'orange'
    else:
        return 'red'

# Add countries to the map
for _, row in merged.iterrows():
    if pd.notna(row['average_intensity']):
        folium.GeoJson(
            row['geometry'],
            style_function=lambda feature, intensity=row['average_intensity']: {
                'fillColor': get_color(intensity),
                'color': 'black',
                'weight': 1,
                'dashArray': '5, 5',
                'fillOpacity': 0.7,
            }
        ).add_to(m)
        folium.Marker(
            location=[row['geometry'].centroid.y, row['geometry'].centroid.x],
            icon=folium.DivIcon(html=f'<div style="font-size: 8pt">{row.name}</div>')
        ).add_to(m)

# Highlight the country with the highest average intensity
highlight_country = merged.loc[max_intensity_country['country']]
folium.GeoJson(
    highlight_country['geometry'],
    style_function=lambda feature: {
        'fillColor': get_color(highlight_country['average_intensity']),
        'color': 'blue',
        'weight': 3,
        'fillOpacity': 0.7,
    }
).add_child(folium.Popup(f"{highlight_country.name}: {highlight_country['average_intensity']}")).add_to(m)

# Save map to HTML file
m.save('average_intensity_map.html')

# If running in a Jupyter notebook, display the map
m
```
